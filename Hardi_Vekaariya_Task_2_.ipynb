**Step 1: Import Required Libraries**

First import Python libraries.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Step 2: Load the Dataset**

```python
df = pd.read_csv("/Unemployment_Rate_upto_11_2020.csv")
```

**step 3: Display Dataset**

```python
print(df.head())
```

**Step 4: Check Dataset Information**

```python
print(df.info())
```

**Step 5: Check Missing Values**

```python
print(df.isnull().sum())
```

**Step 6: Clean the Data**

```python
df = df.dropna()
```

**step 7: Rename Columns**

```python
df.columns = ['State', 'Date', 'Frequency',
              'Estimated Unemployment Rate',
              'Estimated Employed',
              'Labour Participation Rate',
              'Region', 'longitude', 'latitude']
```

**Step 8: Convert Date Column**

```python
df['Date'] = pd.to_datetime(df['Date'])
```

**Step 9: Basic Statistical Analysis**

```python
print(df.describe())
```

**Step 10: Analyze Average Unemployment Rate**

```python
avg_unemployment = df['Estimated Unemployment Rate'].mean()
print(avg_unemployment)
```

**Step 11: State-wise Unemployment Analysis**

```python
state_unemployment = df.groupby('State')['Estimated Unemployment Rate'].mean()

print(state_unemployment)
```

**Step 12: Data Visualization**

A. Line Plot (Trend Over Time)

```python
plt.figure(figsize=(12,6))

sns.lineplot(x='Date',
             y='Estimated Unemployment Rate',
             data=df)

plt.title("Unemployment Rate Over Time")
plt.show()
```

B. Bar Graph (State-wise Analysis)

```python
plt.figure(figsize=(14,7))

sns.barplot(x='State',
            y='Estimated Unemployment Rate',
            data=df)

plt.xticks(rotation=90)

plt.show()
```

C. Heatmap

```python
numeric_df = df.select_dtypes(include=np.number)

sns.heatmap(numeric_df.corr(),
            annot=True,
            cmap='coolwarm')

plt.show()
```

**Step 13: Covid-19 Impact Analysis**

compare both :

Before Covid

During Covid

```python
covid_period = df[df['Date'] >= '2020-03-01']

print(covid_period.head())
```

**Step 14: Draw Covid Impact Graph**

```python
plt.figure(figsize=(12,6))

sns.lineplot(x='Date',
             y='Estimated Unemployment Rate',
             data=covid_period)

plt.title("Unemployment During Covid-19")

plt.show()
```

* Unemployment sharply increased during Covid-19.
* Some states had higher unemployment than others.
* Labour participation decreased during lockdown.
* Employment gradually recovered after restrictions were removed
